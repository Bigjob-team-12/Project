```python
import tensorflow as tf
import numpy as np
import cv2
import os 
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
'''
이 파일은 train 폴더 전, 전 폴더에 있어야함. 
find_folder 매겨변수로 train 폴더를 포함하는 폴더 이름 넣어주면 사용가능.
'''


np.random.seed(15)
folderList = [] 

 
generator = ImageDataGenerator(rescale=1./255,
                                   rotation_range=40, 
                                   shear_range=5.5,  
                                   width_shift_range=0.1,
                                   height_shift_range=0.1,
                                   horizontal_flip=True,
                                   brightness_range=[0.5 ,1.0], 
                                   fill_mode='nearest')

def find_folder(target):
    global path_folder
    path_folder = './' + target +'/' + 'train'
    for count, foldername in enumerate(os.listdir(path_folder)): 
        folderList.append(foldername)    
        
        
        
        
        
find_folder('ours_dog')
print(folderList)
for _ in folderList:
    path = path_folder + '/' + _
    print(path)
    for count, fileName in enumerate(os.listdir(path)): 
        onlyName =  fileName.split('.')[0]
        
        img = load_img(path+'/'+fileName) 
        x = img_to_array(img)
        x = x.reshape((1,) + x.shape)
        i = 0
        for batch in generator.flow(x,save_to_dir=path+'/', save_prefix='aug_'+onlyName, save_format='jpg'):
            i += 1
            print(path+fileName)
            if i > 4:
                i=0
                break
    

```
