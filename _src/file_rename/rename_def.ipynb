```python
import os 
  
def rename(target, tvt): # 파일 번호순으로 이름 바꾸기
    '''
    현재 이 파일은 target > train | valid | test 구조에서 target과 같은 위치에 있어야함 
    target= 현재 ipynb와 같은 경로에 있고, train, valid, test가 들어있는 폴더 
    tvt = train | valid | test 중 하나 선택
    '''
    path_file = './' + target +'/' + tvt
    for count, foldername in enumerate(os.listdir(path_file)): 
        print(count, foldername)
        for count, filename in enumerate(os.listdir(path_file+ '/' +foldername)): 
            print(count, filename)
            #dst ="/" +'new' +str(count) + ".jpg" # 중복된 파일 있을시 실행
            dst ="/" +str(count) + ".jpg"
            src = path_file+ '/'+foldername +'/'+ filename 
            dst = path_file+ '/'+foldername + dst 
            print(src)
            print(dst)
            os.rename(src, dst) 

if __name__ == '__main__': 
    rename('ours_dog', 'test') 
```

```python
import os 
  
def mkdir(target_copy, tvt, target_paste): # 같은이름 폴더 만들기
    '''
    target_copy = class 라벨링 폴더 경로
    tvt = train | valid | test 중 하나 선택
    target_paste = 폴더를 생성할 폴더 경로 지정
    '''
    path_dir_copy =  './' + target_copy 
    path_dir_paste =  './' + target_paste +'/' + tvt

    for count, foldername in enumerate(os.listdir(path_dir_copy)): 
       # print(count, foldername)
       # path = './test/'+foldername
        path = path_dir_paste+'/'+foldername
        print(path)
        os.mkdir(path)
        
if __name__ == '__main__':
    mkdir('dog', 'test', 'dog_test' )
```

```python
import os 
import random

def shuffle(target): # 파일 셔플해서 새로 저장
    '''
    현재 이 파일은 target > class 라벨링 폴더 구조에서 target과 같이 있어야함.
    target= 현재 ipynb와 같은 경로에 있고, class 라벨링 폴더가 안에 있는 폴더

    '''
    list = []  
    count_files = 0 
    path_file = './' + target 
    for count, foldername in enumerate(os.listdir(path_file)): 
        print(count, foldername)
        
        for count, filename in enumerate(os.listdir(path_file+ '/' +foldername)): 
            print(count, filename)
            list.append(filename)
        print(list)
        random.shuffle(list)
        #print(list)
        count_files = 0 
        for _ in list:
            print(_)
            
            #dst ="/" +'shuffled_' +str(count_files) + ".jpg" 
            dst ="/" +'shuf_' +str(count_files) + ".jpg"  # 이름 겹칠때 사용
            src = path_file+ '/'+foldername +'/'+ _ 
            dst = path_file+ '/'+foldername + dst 
            print(src)
            print(dst)
            os.rename(src, dst) 
            count_files = count_files+1
        

        list=[]

if __name__ == '__main__': 
    shuffle('dog') 
```
