## 본 ipynb 파일은 path > category 구조에서 path와 동일한 위치에 있어야 함.

```python
import os
import random
import re
import shutil
```

### dog_breed_color 데이터셋 기초 통계

```python
for i in os.listdir('dogs_breed_color'):
    print(i, len(os.listdir('dogs_breed_color' + '/' + i)))
```

### dogs_breed_color 데이터셋을 dogs_breed 데이터셋으로 카테고리 병합 

```python
for name in os.listdir('dogs_breed_color'):
    print(re.findall(r'([0-9X]+[_][A-Z]+[_]?[A-Z]+[_]?[A-Z]+)', name))
    
    [dst] = re.findall(r'([0-9X]+[_][A-Z]+[_]?[A-Z]+[_]?[A-Z]+)', name)
    dst = 'dogs_breed'  + '/' + dst
    
    for file in os.listdir('dogs_breed_color' + '/' + name):
        src = 'dogs_breed_color' + '/' + name + '/' + file
        shutil.copy(src, dst)
```

### 랜덤 순서로 이름 바꾸기

```python
def shuffle_rename(path): # path는 카테고리 폴더들이 들어 있는 상위 폴더 이름

    for breed in os.listdir(path):
        total = len(os.listdir(path + '/' + breed)) # 각 품종 데이터 수
        name_list = [f'{i:0>4}.jpg' for i in range(total)] # '0000.jpg'부터 시작해 데이터 수만큼 이름 리스트 생성
        random.shuffle(name_list) # 이름 리스트 셔플
        
        for file in os.listdir(path + '/' + breed):
            new_name = name_list.pop(0) # 이름 리스트에서 하나씩 꺼내 기존 이름 변경
            src = path + '/' + breed + '/' + file
            dst = path + '/' + breed + '/' + new_name
            print(src, dst)
            os.rename(src, dst)
```

```python
shuffle_rename('dogs_breed')
```

### 3-way hold-out

```python
def tvt_split(path, valid=.2, test=.2, same=True): # same 옵션은 클래스별 데이터 개수 통일 여부
    
    os.mkdir('new ' + path) # 현재 경로에 train, valid, test로 나뉜 새 폴더 생성
    os.mkdir('new ' + path + '/' + 'train')
    os.mkdir('new ' + path + '/' + 'valid')
    os.mkdir('new ' + path + '/' + 'test')
    
    if same == True:
        cnt_list = [len(os.listdir(path + '/' + breed)) for breed in os.listdir(path)]
        num_data = min(cnt_list) # 데이터 가장 적은 클래스의 데이터 수
        
        num_test = round(num_data * test) # 성능 평가 시 test set을 확인하므로 앞쪽 넘버링 배치
        num_valid = round(num_data * valid)
        
        for breed in os.listdir(path):
            print(breed)
            os.mkdir('new ' + path + '/' + 'train' + '/' + breed)
            os.mkdir('new ' + path + '/' + 'valid' + '/' + breed)
            os.mkdir('new ' + path + '/' + 'test' + '/' + breed)
            
            for _ in os.listdir(path + '/' + breed)[:num_test]: # test set 복사
                src = path + '/' + breed + '/' + _
                dst = 'new ' + path + '/' + 'test' + '/' + breed
                shutil.copy(src, dst)
            for _ in os.listdir(path + '/' + breed)[num_test:num_test+num_valid]: # valid set 복사
                src = path + '/' + breed + '/' + _
                dst = 'new ' + path + '/' + 'valid' + '/' + breed
                shutil.copy(src, dst)
            for _ in os.listdir(path + '/' + breed)[num_test+num_valid:num_data]: # train set 복사
                src = path + '/' + breed + '/' + _
                dst = 'new ' + path + '/' + 'train' + '/' + breed
                shutil.copy(src, dst)
                
    elif same == False:
        for breed in os.listdir(path):
            print(breed)
            num_data = len(os.listdir(path + '/' + breed))
            
            num_test = round(num_data * test)
            num_valid = round(num_data * valid)
                        
            os.mkdir('new ' + path + '/' + 'train' + '/' + breed)
            os.mkdir('new ' + path + '/' + 'valid' + '/' + breed)
            os.mkdir('new ' + path + '/' + 'test' + '/' + breed)
            
            for _ in os.listdir(path + '/' + breed)[:num_test]:
                src = path + '/' + breed + '/' + _
                dst = 'new ' + path + '/' + 'test' + '/' + breed
                shutil.copy(src, dst)
            for _ in os.listdir(path + '/' + breed)[num_test:num_test+num_valid]:
                src = path + '/' + breed + '/' + _
                dst = 'new ' + path + '/' + 'valid' + '/' + breed
                shutil.copy(src, dst)
            for _ in os.listdir(path + '/' + breed)[num_test+num_valid:]:
                src = path + '/' + breed + '/' + _
                dst = 'new ' + path + '/' + 'train' + '/' + breed
                shutil.copy(src, dst)
```

```python
tvt_split('dogs_breed')
```
