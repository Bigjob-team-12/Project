```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re
pd.set_option('display.max_rows', None)
```

```python
pattern = r"[A-Z]+"

all_list = os.listdir('dogs_breed_0820')
ttv = []
breed = []
count = []
for cate in all_list:
    folder_list = os.listdir(os.path.join('dogs_breed_0820', cate))
    for folder in folder_list:
        ttv.append(cate)
        breed.append(' '.join(re.findall(pattern, folder[1:])))
        count.append(len(os.listdir(os.path.join('dogs_breed_0820', cate, folder))))
```

```python
# to dataframe
df = pd.DataFrame([ttv, breed, count]).T
df.columns = ['set', 'breed', 'count']
df
```

```python
# 품종별 개수
df.sort_values(by=['breed', 'set'], axis=0).set_index(['breed', 'set'])
```

```python
# 용도별 개수
df.sort_values(by=['set', 'breed'], axis=0).set_index(['set', 'breed'])
```

```python
# 비율
df['set_in_breed_ratio'] = df['count']/df.groupby('breed')['count'].sum().loc[df['breed']].values*100 # 품종 내 set 비중
df['breed_in_set_ratio'] = df['count']/df.groupby('set')['count'].sum().loc[df['set']].values*100 # set 내 품종 비중

# 품종별
df.sort_values(by=['breed', 'set'], axis=0).set_index(['breed', 'set'])
```

```python
# 용도별
df.sort_values(by=['set', 'breed'], axis=0).set_index(['set', 'breed'])
```

```python
# 전체 품종별 비율
breedratio = pd.DataFrame(df.groupby('breed')['count'].sum()).reset_index()
breedratio['ratio'] = breedratio['count']/breedratio['count'].sum()*100
breedratio.sort_values(by=['ratio'], ascending=False, axis=0)
```

```python
# 전체 set 비율
setratio = pd.DataFrame(df.groupby('set')['count'].sum()).reset_index()
setratio['ratio'] = setratio['count']/setratio['count'].sum()*100
setratio.set_index('set')
```

```python
# donut graph
plt.figure(figsize=(40, 40))
plt.rc('font', family='NanumGothic')
plt.rcParams['font.size'] = 15
fig, ax = plt.subplots()
size = 2

cmap = plt.get_cmap("Pastel2")
outer_colors = cmap(np.arange(8))
inner_colors = cmap([1, 0, 2])

dfdf = df.sort_values(by=['breed', 'set'], axis=0)
temp = pd.DataFrame(dfdf.groupby('breed')['count'].sum()).reset_index()

ax.pie(dfdf['count'], labels=dfdf['set'], radius=5-size, colors=inner_colors, wedgeprops=dict(width=size, edgecolor='w'), labeldistance=1, startangle=90)
ax.pie(temp['count'], labels=temp['breed'], radius=5, colors=outer_colors, wedgeprops=dict(width=size, edgecolor='w'), labeldistance=1.05, startangle=90)

plt.show()
```

```python
folder_list = os.listdir('re_id_dataset')

pattern1 = r"[A-Z]+"
pattern2 = r"[a-z]+"
all_list = os.listdir('re_id_dataset')

ttv = []
breed = []
color = []
count = []

for cate in all_list:
    folder_list = os.listdir(os.path.join('re_id_dataset', cate))
    for folder in folder_list:
        ttv.append(cate)
        breed.append(' '.join(re.findall(pattern1, folder[1:])))
        color.append(' '.join(re.findall(pattern2, folder[1:])))
        count.append(len(os.listdir(os.path.join('re_id_dataset', cate, folder))))
```

```python
# to dataframe
df = pd.DataFrame([breed, color, ttv, count]).T
df.columns = ['breed', 'color', 'set', 'count']
df = df.sort_values(by=['breed', 'color', 'set'], axis=0).reset_index().drop('index', axis=1)
```

```python
df
```

```python
# 비율
df2 = pd.DataFrame(df.groupby(['breed', 'color']).sum()['count'])
df2 = df2.reset_index()
df2['color_in_breed_ratio'] = df2['count']/df2.groupby('breed')['count'].sum().loc[df2['breed']].values*100 # 품종 내 color 비중
df2['breed_in_color_ratio'] = df2['count']/df2.groupby('color')['count'].sum().loc[df2['color']].values*100 # color 내 품종 비중
df2.set_index(['breed', 'color'])
```

```python
# 전체 품종별 비율
breedratio = pd.DataFrame(df.groupby('breed')['count'].sum()).reset_index()
breedratio['ratio'] = breedratio['count']/breedratio['count'].sum()*100
breedratio = breedratio.sort_values(by=['ratio'], ascending=False, axis=0).reset_index().drop('index', axis=1)
breedratio
```

```python
# 전체 color 비율
colorratio = pd.DataFrame(df.groupby('color')['count'].sum()).reset_index()
colorratio['ratio'] = colorratio['count']/colorratio['count'].sum()*100
colorratio.sort_values(by=['ratio'], ascending=False, axis=0)
```

```python
# donut graph
plt.figure(figsize=(40, 40))
plt.rc('font', family='NanumGothic')
plt.rcParams['font.size'] = 15
fig, ax = plt.subplots()
size = 1.5

cmap = plt.get_cmap("Pastel2")
outer_colors = cmap(np.arange(8))
inner_colors = cmap(np.arange(8))
tri_colors = cmap(np.arange(3))

# 각 color 내 set
ax.pie(pd.DataFrame(breedratio['breed']).merge(df)['count'], radius=5-(2*size), colors=tri_colors, wedgeprops=dict(width=size, edgecolor='w'), startangle=90)

# 각 품종 내 color
ax.pie(pd.DataFrame(breedratio['breed']).merge(pd.DataFrame(df.groupby(['breed', 'color'])['count'].sum()).reset_index())['count'], labels=pd.DataFrame(breedratio['breed']).merge(pd.DataFrame(df.groupby(['breed', 'color'])['count'].sum()).reset_index())['color'], radius=5-size, colors=inner_colors, wedgeprops=dict(width=size, edgecolor='w'), labeldistance=0.9, startangle=90)

# 전체 품종
ax.pie(pd.DataFrame(breedratio['breed']).merge(pd.DataFrame(df.groupby('breed')['count'].sum()).reset_index())['count'], labels=pd.DataFrame(breedratio['breed']).merge(pd.DataFrame(df.groupby('breed')['count'].sum()).reset_index())['breed'], radius=5, colors=outer_colors, wedgeprops=dict(width=size, edgecolor='w'), labeldistance=1.05, startangle=90)

plt.show()
```
